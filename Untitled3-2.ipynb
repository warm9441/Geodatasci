```python
# นำข้อมูลวันและเวลามาต่อกัน
Fire['ACQ_TIME']=Fire['ACQ_TIME']/100
Fire['ACQ_TIME']=Fire['ACQ_TIME'].astype(str)
Fire['ACQ_TIME']
```

```python
Fire['ACQ_TIME']=Fire['ACQ_TIME'].str.replace('.',':',regex=True)
Fire['DATE&TIME']=Fire['ACQ_DATE'].astype(str)+' '+Fire['ACQ_TIME']
Fire['DATE&TIME']=pd.to_datetime(Fire['DATE&TIME'])
Fire['DATE&TIME']
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fire[Fire['DATE&TIME']<=(Fire['DATE&TIME'].iloc[0]+pd.Timedelta(days=5))]
```

```python
Fire['DT']=Fire['DATE&TIME']-Fire['DATE&TIME'][0]
Fire['DD']=(Fire['DT'].dt.total_seconds())/(60*60*24)
Fire
```

```python
Fire_5=Fire[Fire['DATE&TIME']<=(Fire['DATE&TIME'].iloc[0]+pd.Timedelta(days=5))]
```

```python
# Plot 3D
fig=plt.figure()
axes=fig.add_subplot(111,projection='3d')
X=Fire_5['LONGITUDE']
Y=Fire_5['LATITUDE']
Z=Fire_5['DD']
axes.scatter(X,Y,Z)
axes.set_xlabel('Longitude')
axes.set_ylabel('Latitude')
axes.set_zlabel('Days')
plt.show()
```
